```python
import numpy as np
```

```python
x= np.linspace(-10,10,100)
```

```python
y= 5+4*x-3*x**2+np.random.random(100)*100
```

```python
from matplotlib import pyplot as plt
```

```python
plt.scatter(x,y)
```

```python
X=x.reshape(-1,1)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
def compute_model_train_test_values(X,y,test_size=0.25):
    X_train,X_test,y_train,y_test=train_test_split(X,y,
                                                   test_size=test_size)
    model =LinearRegression()
    model.fit(X_train,y_train)
    y_pred_train=model.predict(X_train)
    rmse_train=mean_squared_error(y_train,y_pred_train)**0.5
    r2_train=model.score(X_train,y_train)
    
    y_pred_test=model.predict(X_test)
    rmse_test=mean_squared_error(y_test,y_pred_test)**0.5
    r2_test=model.score(X_test,y_test)
    return {
            "testing accuracy":r2_test,
            "training accuracy":r2_train,
            "rmse_test":rmse_test,
            "rmse_train":rmse_train,
            "intercept":model.intercept_,
            "coeff":model.coef_,
            "model":model
           }
```

```python
result =compute_model_train_test_values(X,y)
```

```python
result
```

```python
y_pred=result["model"].predict(X)
```

```python
y_pred
```

```python
y
```

```python
plt.scatter(x,y)
plt.plot(x,y_pred)
```
