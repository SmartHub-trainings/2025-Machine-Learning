## Grid Search CV

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.datasets import load_breast_cancer
X,y= load_breast_cancer(return_X_y=True)
```

```python
knn_model = KNeighborsClassifier()
n_neighbors=range(3,20)
p = range(1,10)
```

```python
from sklearn.model_selection import (GridSearchCV,
                                     train_test_split,
                                    RandomizedSearchCV)
```

```python
## baseline model
knn_model = KNeighborsClassifier()
X_train,X_test,y_train,y_test = train_test_split(X,y)
knn_model.fit(X_train,y_train)
print(f"Training accuracy {knn_model.score(X_train,y_train)}")
print(f"Testing accuracy {knn_model.score(X_test,y_test)}")
```

```python
params ={
    "n_neighbors":n_neighbors,
    "p":p, 
}

grid_model=GridSearchCV(knn_model,param_grid=params)
```

```python
%%time
grid_model.fit(X,y)
```

```python
grid_model.best_estimator_
```

```python
grid_model.best_params_
```

```python
grid_model.best_score_
```

## Randomized Search CV

```python
randomized_model=RandomizedSearchCV(knn_model,param_distributions=params,n_iter=50)
```

```python
%%time
randomized_model.fit(X,y)
```

```python
randomized_model.best_score_
```

```python
randomized_model.best_params_
```
