## Multiple linear regression

```python
from sklearn.datasets import load_boston
```

```python
bundle=load_boston()
```

```python
print(bundle.DESCR)
```

```python
bundle.feature_names
```

```python
y=bundle.target
```

```python
X=bundle.data
```

```python
X.shape
```

```python
X_rm=X[:,5].reshape(-1,1)
```

```python
X_rm
```

```python
def compute_model_values(X,y):
    model =LinearRegression()
    model.fit(X,y)
    y_pred=model.predict(X)
    rmse=mean_squared_error(y,y_pred)**0.5
    r2=r2_score(y,y_pred)
    return {"r2":r2,"rmse":rmse,"y_pred":y_pred,
            "intercept":model.intercept_,
            "coeff":model.coef_
           }
```

```python
compute_model_values(X_rm,y)
```

```python
X_comb1=X[:,[1,5]]
```

```python
compute_model_values(X_comb1,y)
```

```python
X_comb2=X[:,[0,1,5]]
```

```python
compute_model_values(X_comb2,y)
```
