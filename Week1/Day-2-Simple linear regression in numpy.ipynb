```python
import numpy as np
```

```python
np.random.seed(42)
heights=np.random.randint(3,8,30)
weights=4+20*heights+np.random.random(30)*20
```

```python
#beta_0=4 and beta_1=20
```

```python
heights
```

```python
weights
```

```python
from matplotlib import pyplot as plt
```

```python
plt.scatter(heights,weights)
```

```python
n= len(weights)
```

```python
sum_x=heights.sum()
sum_y=weights.sum()
sum_x_sq=(heights**2).sum()
sum_xy = (weights*heights).sum()
```

```python
beta_1=(n*sum_xy-sum_x*sum_y)/(n*sum_x_sq-sum_x**2)
```

```python
beta_1
```

```python
beta_0 = sum_y/n-beta_1*sum_x/n
```

```python
beta_0
```

```python
weights_pred=beta_0+beta_1*heights
```

```python
weights_pred
```

```python
weights
```

```python
residuals = weights-weights_pred
```

```python
residuals
```

```python
residuals.sum()
```

```python
mae = (np.abs(residuals)).mean()
```

```python
mae
```

```python
rmse= ((residuals**2).mean())**0.5
```
